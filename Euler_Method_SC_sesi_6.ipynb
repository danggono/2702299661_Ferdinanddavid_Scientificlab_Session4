Euler Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-poster')
%matplotlib inline

#define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step  size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
  s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \ for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the differential equation y' + 4y = x^2
def f(x, y):
    return x**2 - 4*y

# Parameters
h = 0.01  # Step size
t = np.arange(0, 1 + h, h)  # Numerical grid
y0 = 1  # Initial Condition

# Explicit Euler Method
y = np.zeros(len(t))
y[0] = y0

for i in range(0, len(t) - 1):
    y[i + 1] = y[i] + h * f(t[i], y[i])

# Exact Solution
exact_y = (1/16) * (1 - np.exp(-4*t)) + (t**3) / 12 - (t**2) / 8 + t / 32

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(t, y, 'b--', label='Approximate')
plt.plot(t, exact_y, 'g', label='Exact')
plt.title("Approximate and Exact Solution \n for the ODE $y' + 4y = x^2$")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster') # Use the updated style name
%matplotlib inline

# Define parameters for the new ODE
g = lambda t, y: np.sin(y) # New ODE
h = 0.1 # Step size
x = np.arange(0, 0.5 + h, h) # Numerical grid
y0 = 1 # Initial Condition

# Explicit Euler Method for the new ODE
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * g(x[i], y[i])
    y[i + 1] = np.round(y[i + 1], 4)  # Keep four decimal places

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo--', label='Approximate Solution')
plt.title('Approximate Solution for y\' = sin(y) with Explicit Euler Method')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
